```python
from typing import Tuple

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

# For the Heisenberg "tetrahedron"

```python
def f(x, y, n):
    "The surface polynomial to be used."
    return n - x - y + x*y

def make_tet_surface(n: int, fineness: int) \
   -> Tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Given n and a fineness, generate a plot of the function f over
    the 2D base of the nth dilate of the unit tetrahedron.
    
    (that's a stupid way complicated way to write something that
    I'm not sure how to write coherently right now.)
    """
    length = fineness + 1
    x, y = np.mgrid[0.0:n:length*1j, 0.0:n:length*1j]
    in_region = y < n + EPSILON - x
    x, y = x[in_region], y[in_region]
    z = f(x, y, n)
    z[z < 0] = 0.0
    return x, y, z
```

```python
N = 10
FINENESS = 20
EPSILON = 1/(FINENESS**3)

SUBPLOT = 111
fig = plt.figure()
ax = fig.add_subplot(SUBPLOT, projection='3d')
x, y, z = make_tet_surface(N, FINENESS)
ax.plot_trisurf(x, y, z, edgecolor='none')
plt.show()
```

# For the Heisenberg vomit square wave thing

```python
def vomit_roof(x, y, n):
    return np.select(
        (
            x+y <= n,
            x+y > n
        ),
        (
            x*y,
            - x - y + x*y + n
        )
    )

def vomit_floor(x, y, n):
    return np.select(
        (
            x + y <= n,
            x + y > n
        ),
        (
            np.zeros(x.shape), 
            x**2 + 2*x*y + y**2 + n**2 + n - (2*n + 1)*(x + y)
        )
    )

def make_vomit_square_roof(n: int, fineness: int) \
   -> Tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Given n and a fineness, generate a plot of the nth Heisenberg
    vomit square roof-surface.
    """
    length = fineness + 1
    x, y = np.mgrid[0.0:n:length*1j, 0.0:n:length*1j]
    
    z = vomit_roof(x, y, n)
    # z[z < 0] = 0.0
    
    return x, y, z

def make_vomit_square_floor(n: int, fineness: int) \
   -> Tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Given n and a fineness, generate a plot of the nth Heisenberg
    vomit square floor-surface.
    """
    length = fineness + 1
    x, y = np.mgrid[0.0:n:length*1j, 0.0:n:length*1j]
    
    z = vomit_floor(x, y, n)
    z[z < 0] = 0.0
    
    return x, y, z
```

```python
N = 4
FINENESS = 20
EPSILON = 1/(FINENESS**3)

fig = plt.figure()
ax1 = fig.add_subplot(121, projection='3d')
# Compute, then unroll into 1D arrays for plotting purposes
x, y, z = make_vomit_square_roof(N, FINENESS)
x, y, z = x.reshape(-1), y.reshape(-1), z.reshape(-1)
ax1.plot_trisurf(x, y, z, edgecolor='none')

ax2 = fig.add_subplot(122, projection='3d')
x, y, z = make_vomit_square_floor(N, FINENESS)
x, y, z = x.reshape(-1), y.reshape(-1), z.reshape(-1)
ax2.plot_trisurf(x, y, z, edgecolor='none')

plt.show()
```
