```python
from typing import Tuple

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
# Parameters
R = 100.0
stride = 1.0

# Make grid and filter to points in the ball
x, y = np.mgrid[-R:R:stride, -R:R:stride]
in_region = x*x + y*y < (R*R + 1.0)
x, y = x[in_region], y[in_region]

# Now make the top and bottom of the ball
z_top = -x*x - y*y + R*R
z_bot = -z_top

# Plot
x_p = np.concatenate((x, x))
y_p = np.concatenate((y, y))
z_p = np.concatenate((z_top, z_bot))

SUBPLOT = 111
fig = plt.figure()
ax = fig.add_subplot(SUBPLOT, projection='3d')
# ax.plot_trisurf(x, y, z_bot, edgecolor='none')
ax.plot_trisurf(x, y, z_top, edgecolor='none')
plt.show()
```
