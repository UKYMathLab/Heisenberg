```python
from pathlib import Path
import sys
import itertools as it

import numpy as np
from tqdm import tqdm
import pptk

# custom path so I can import from src folder of project directory
sys.path.append("E:\\User\\Documents\\Projects\\Heisenberg\\src")
from HeisenbergCompute import HeisenbergVectors
from StandardCompute import StandardVectors
from utils import point_cloud
import drivers
```

```python
basis_vectors =np.array([[0, 0, 0],
                         [1, 0, 0],
                         [0, 1, 0],
                         [0, 0, 1]], dtype=int)
```

```python
# check if k-th dilation of Heisenberg polytope:
#     (1) contains an m-th dilate tetrahedron
#     (2) is contained in an n-th dilate tetrahedron
num_sums = 2


# start with initial dilation
Pm = StandardVectors(basis_vectors, num_sums)
Pk = HeisenbergVectors(basis_vectors, num_sums)
Pn = StandardVectors(basis_vectors, num_sums)
polytopes = (Pm, Pk, Pn)
```

```python
def check_containment(polytopes: list) -> tuple:
    containments = (True, True)
    for i in range(len(polytopes)-1):
        for row in polytopes[i].unique_permutations.tolist():
            if row not in polytopes[i+1].unique_permutations.tolist():
                containments[i] = False
    return containments
```

```python
both_contained = all(check_containment(polytopes))
while both_contained:
    num_sums += 1
    num_sums_range = range(2, num_sums)
    mkn = it.product(num_sums_range, repeat=3)
    
    print(num_sums_range)
    for m, k, n in mkn:
        Pm = StandardVectors(basis_vectors, m)
        Pk = HeisenbergVectors(basis_vectors, k)
        Pn = StandardVectors(basis_vectors, n)
        polytopes = (Pm, Pk, Pn)
        
        both_contained = all(check_containment(polytopes))
        
        if both_contained:
            break
print(f"m={m}\nk={k}\nn={n}")
```
